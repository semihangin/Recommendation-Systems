```python
# tüm işlemlerin fonksiyonlaştırılması
def create_user_movie_df():
    import pandas as pd
    movie = pd.read_csv(filepath_movie)
    rating = pd.read_csv(filepath_rating)
    df = movie.merge(rating, how="left", on="movieId")
    comment_counts = pd.DataFrame(df["title"].value_counts())
    rare_movies = comment_counts[comment_counts["title"] <= 1000].index
    common_movies = df[~df["title"].isin(rare_movies)]
    user_movie_df = common_movies.pivot_table(index=["userId"], columns=["title"], values="rating")
    return user_movie_df

user_movie_df = create_user_movie_df()
```

Öneri Yapılacak Kullanıcının İzlediği Filmlerin Belirlenmesi

```python
# Rastgele bir kullanıcı id'si seçiyoruz.
random_user = 108170
```

```python
# Seçilen kullanıcıya ait gözlem birimlerinden oluşan random_user_df adında yeni bir dataframe oluşturuyoruz.
random_user_df = user_movie_df[user_movie_df.index == random_user]
random_user_df.head()
```

```python
# Seçilen kullanıcının oy kullandığı filmleri movies_watched adında bir listeye atıyoruz.
movies_watched = random_user_df.columns[random_user_df.notna().any()].to_list()
movies_watched
```

```python
movie.columns[movie.notna().any()].to_list()
```

Aynı Filmleri İzleyen Diğer Kullanıcıların Verisine ve Id'lerine Erişmek

```python
# Seçilen kullanıcının izlediği fimlere ait sutunları user_movie_df'ten seçiniz ve movies_watched_df adında yeni bir dataframe oluşturuyoruz.
movies_watched_df = user_movie_df[movies_watched]
movies_watched_df.shape
```

```python
# Herbir kullancının seçili user'in izlediği filmlerin kaçını izlediği bilgisini taşıyan user_movie_count adında yeni bir dataframe oluşturuyoruz.
# Ve yeni bir df oluşturuyoruz.
user_movie_count = movies_watched_df.T.notnull().sum()
user_movie_count = user_movie_count.reset_index()
user_movie_count.columns = ["userId", "movie_count"]
user_movie_count.head(5)
```

```python
# Seçilen kullanıcının oy verdiği filmlerin yüzde 60 ve üstünü izleyenleri benzer kullanıcılar olarak görüyoruz.
# Bu kullanıcıların id’lerinden users_same_movies adında bir liste oluşturuyoruz.
perc = len(movies_watched) * 60 / 100
users_same_movies = user_movie_count[user_movie_count["movie_count"] > perc]["userId"]
len(users_same_movies)
```

Öneri Yapılacak Kullanıcı ile En Benzer Kullanıcıların Belirlenmesi

```python
# user_same_movies listesi içerisindeki seçili user ile benzerlik gösteren kullanıcıların id’lerinin bulunacağı şekilde movies_watched_df dataframe’ini filtreliyoruz.
final_df = movies_watched_df[movies_watched_df.index.isin(users_same_movies)]
final_df.head()
```

```python
# Kullanıcıların birbirleri ile olan korelasyonlarının bulunacağı yeni bir corr_df dataframe’i oluşturunuz.
corr_df = final_df.T.corr().unstack().sort_values()
corr_df = pd.DataFrame(corr_df, columns=["corr"])
corr_df.index.names = ['user_id_1', 'user_id_2']
corr_df = corr_df.reset_index()
```

```python
corr_df[corr_df["user_id_1"] == random_user]
```

```python
# Seçili kullanıcı ile yüksek korelasyona sahip (0.65’in üzerinde olan) kullanıcıları filtreleyerek top_users adında yeni bir dataframe oluşturuyoruz.
top_users = corr_df[(corr_df["user_id_1"] == random_user) & (corr_df["corr"] >= 0.65)][["user_id_2", "corr"]].reset_index(drop=True)
top_users = top_users.sort_values(by='corr', ascending=False)
top_users.rename(columns={"user_id_2": "userId"}, inplace=True)
top_users.head()
```

```python
# top_users dataframe’ini rating veri seti ile merge edinyoruz ve top_users_ratings adında yeni bir dataframe oluşturuyoruz.
top_users_ratings = top_users.merge(rating[["userId", "movieId", "rating"]], how='inner')
top_users_ratings = top_users_ratings[top_users_ratings["userId"] != random_user]
top_users_ratings["userId"].unique()
```

```python
top_users_ratings.head()
```

Weighted Average Recommendation Score'un Hesaplanması ve İlk 5 Filmin Tutulması

```python
# Her bir kullanıcının corr ve rating değerlerinin çarpımından oluşan weighted_rating adında yeni bir değişken oluşturuyoruz
top_users_ratings['weighted_rating'] = top_users_ratings['corr'] * top_users_ratings['rating']
top_users_ratings.head()
```

```python
# Film id’sive her bir filme ait tüm kullanıcıların weighted rating’lerinin ortalama değerini içeren recommendation_df adında yeni bir
# dataframe oluşturuyoruz.
recommendation_df = top_users_ratings.groupby('movieId').agg({"weighted_rating": "mean"})
recommendation_df = recommendation_df.reset_index()
recommendation_df.head()
```

```python
# recommendation_df içerisinde weighted rating'i 3.5'ten büyük olan filmleri seçiniz ve weighted rating’e göre sıralıyoruz.
# İlk 5 gözlemi movies_to_be_recommend olarak kaydediyoruz.
recommendation_df[recommendation_df["weighted_rating"] > 3.5]
movies_to_be_recommend = recommendation_df[recommendation_df["weighted_rating"] > 3.5].sort_values("weighted_rating", ascending=False)
```

```python
# Tavsiye edilen 5 filmin isimleri
movies_to_be_recommend.merge(movie[["movieId", "title"]])["title"][:5]
```

Item-Based Recommendation

```python
user = 108170
```

```python
movie = pd.read_csv(filepath_movie)
rating = pd.read_csv(filepath_rating)
```

```python
# Öneri yapılacak kullanıcının 5 puan verdiği filmlerden puanı en güncel olan filmin id'sinin alıyoruz.
movie_id = rating[(rating["userId"] == user) & (rating["rating"] == 5.0)].sort_values(by="timestamp", ascending=False)["movieId"][0:1].values[0]
```

```python
# User based recommendation bölümünde oluşturulan user_movie_df dataframe’ini seçilen film id’sine göre filtreliyoruz.
movie[movie["movieId"] == movie_id]["title"].values[0]
movie_df = user_movie_df[movie[movie["movieId"] == movie_id]["title"].values[0]]
movie_df.head()
```

```python
# Filtrelenen dataframe’i kullanarak seçili filmle diğer filmlerin korelasyonunu buluyoruz ve sıralıyoruz.
user_movie_df.corrwith(movie_df).sort_values(ascending=False).head(10)
```

```python
# Son iki adımı uygulayan fonksiyon
def item_based_recommender(movie_name, user_movie_df):
    movie = user_movie_df[movie_name]
    return user_movie_df.corrwith(movie).sort_values(ascending=False).head(10)
```

```python
movies_from_item_based = item_based_recommender(movie[movie["movieId"] == movie_id]["title"].values[0], user_movie_df)
```
